```python
crop_avg = ay_df.groupby(
    'crop_type'
)['yield_ton_per_hectare'].mean()

soil_avg = ay_df.groupby(
    'soil_type'
)['yield_ton_per_hectare'].mean()

print(crop_avg)
print(soil_avg)

print("Highest Yield Crop:")
print(crop_avg.idxmax(), crop_avg.max())

print("Highest Yield Soil:")
print(soil_avg.idxmax(), soil_avg.max())
```

```python
df_encoded = pd.get_dummies(
    ay_df,
    columns=['crop_type','soil_type'],
    drop_first=True
)

print(df_encoded.head())
```

```python
X = df_encoded.drop(
    'yield_ton_per_hectare',
    axis=1
)

y = df_encoded['yield_ton_per_hectare']

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print("Intercept:")
print(model.intercept_)

coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coef_df)

print("\nHighest Positive Coefficient:")
print(
    coef_df.sort_values(
        by='Coefficient',
        ascending=False
    ).head(1)
)
```
